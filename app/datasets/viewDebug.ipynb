```python
# url1 = "./raw/03_windows_dataset_tobs60_tpred30_tlead10.parquet"
url1 = "./processed/03_windows_dataset_tobs60_tpred30_tlead10_indexed.parquet"
# Import a parquet file as a pandas DataFrame
df = pd.read_parquet(url1)
```

```python
# Calcular longitudes de los arrays en observation_Events y prediction_Events
df['len_observation'] = df['observation_events'].apply(len)
df['len_prediction'] = df['prediction_events'].apply(len)

# Estadísticas para observation_Events
print("=== observation_Events ===")
print(f"Longitud media: {df['len_observation'].mean():.2f}")
print(f"Longitud máxima: {df['len_observation'].max()}")
print(f"Longitud mínima: {df['len_observation'].min()}")

print("\n=== prediction_Events ===")
print(f"Longitud media: {df['len_prediction'].mean():.2f}")
print(f"Longitud máxima: {df['len_prediction'].max()}")
print(f"Longitud mínima: {df['len_prediction'].min()}")

# Distribución de longitudes
print("\n=== Distribución de longitudes ===")
print("\nobservation_Events:")
print(df['len_observation'].value_counts().sort_index())
print("\nprediction_Events:")
print(df['len_prediction'].value_counts().sort_index())
```

```python
# Encontrar las filas con los arrays más largos
idx_max_observation = df['len_observation'].idxmax()
idx_max_prediction = df['len_prediction'].idxmax()

print("=== Fila con observation_events más largo ===")
print(f"Índice: {idx_max_observation}")
print(f"Longitud: {df.loc[idx_max_observation, 'len_observation']}")
print(f"\nFila completa:")
print(df.loc[idx_max_observation])
print(f"\nContenido de observation_events:")
print(df.loc[idx_max_observation, 'observation_events'])

print("\n" + "="*50 + "\n")

print("=== Fila con prediction_events más largo ===")
print(f"Índice: {idx_max_prediction}")
print(f"Longitud: {df.loc[idx_max_prediction, 'len_prediction']}")
print(f"\nFila completa:")
print(df.loc[idx_max_prediction])
print(f"\nContenido de prediction_events:")
print(df.loc[idx_max_prediction, 'prediction_events'])
```
